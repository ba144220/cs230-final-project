```python
def create_bar_chart(x, y, title):
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.set_xlabel('Accuracy')
    plt.barh(y=y, width=x)
    plt.title(title)
    plt.savefig('{}.png'.format(title), dpi=300, bbox_inches='tight')
    plt.show()
```

```python
create_bar_chart(self_generated_accuracy, models, "Self-Generated Data Overall Accuracy")
```

```python
create_bar_chart(wtq_accuracy, models, "WikiTableQuestions Overall Accuracy")
```

```python
create_bar_chart(overall_accuracy, models, "Overall Accuracy")
```
